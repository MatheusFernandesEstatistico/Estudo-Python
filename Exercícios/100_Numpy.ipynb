#### 22. Normalizar uma matriz aleatória 5x5 (★☆☆)

````python
matriz_22 = np.random.randint(0, 100, size=(5, 5))
print("Matriz original:")
print(matriz_22)
````

22.1 Normalização Min-Max (escala [0, 1])

````python
min_val = matriz_22.min()
max_val = matriz_22.max()
matriz_norm = (matriz_22 - min_val) / (max_val - min_val)

print("Normalizada [0, 1]:")
print(matriz_norm)
````

22.2 Normalização Min-Max com intervalo customizado [a, b]

````python
a, b = -1, 1
matriz_norm = a + (matriz_22 - matriz_22.min()) * (b - a) / (matriz_22.max() - matriz_22.min())

print("Normalizada [-1, 1]:")
print(matriz_norm)
````

22.3 Normalização Z-score (média 0, desvio padrão 1)

````python
matriz_norm = (matriz_22 - matriz_22.mean()) / matriz_22.std()

print("Z-score (média 0, std 1):")
print(matriz_norm)
print(f"Média: {matriz_norm.mean():.2f}")
print(f"Desvio padrão: {matriz_norm.std():.2f}")
````

22.4 Normalização por linha (cada linha soma 1)

````python
matriz_norm = matriz_22 / matriz_22.sum(axis=1, keepdims=True)

print("Normalizada por linha (soma = 1):")
print(matriz_norm)
print("Soma das linhas:", matriz_norm.sum(axis=1))
````

22.5 Normalização por coluna

````python
matriz_norm = matriz_22 / matriz_22.sum(axis=0, keepdims=True)

print("Normalizada por coluna (soma = 1):")
print(matriz_norm)
print("Soma das colunas:", matriz_norm.sum(axis=0))
````

22.6 Normalização L2 (norma euclidiana = 1)

````python
from numpy.linalg import norm

matriz_norm = matriz_22 / norm(matriz_22)

print("Normalizada L2:")
print(matriz_norm)
print(f"Norma L2: {norm(matriz_norm):.2f}")
````

#### 23. Crie um tipo de dados personalizado (dtype) que descreva uma cor como quatro bytes sem sinal (RGBA) (★☆☆)

````python
cor_rgba = np.dtype([
    ('R', np.uint8),  # Red: 0-255
    ('G', np.uint8),  # Green: 0-255
    ('B', np.uint8),  # Blue: 0-255
    ('A', np.uint8)   # Alpha: 0-255
])

print(cor_rgba)
````

````python
vermelho = np.array((255, 0, 0, 255), dtype=cor_rgba)
print(vermelho)  # (255, 0, 0, 255)

print(f"R: {vermelho['R']}")
print(f"G: {vermelho['G']}")
print(f"B: {vermelho['B']}")
print(f"A: {vermelho['A']}")
````

````python
cores = np.array([
    (255, 0, 0, 255),     # Vermelho
    (0, 255, 0, 255),     # Verde
    (0, 0, 255, 255),     # Azul
    (0, 0, 0, 0),         # Transparente
    (255, 255, 0, 128),   # Amarelo semi-transparente
    (128, 0, 128, 255),   # Roxo
], dtype=cor_rgba)

print("Paleta de cores:")
print(cores)
````

````python
print("Canal Red:", cores['R'])    # [255   0   0   0 255 128]
print("Canal Green:", cores['G'])  # [  0 255   0   0 255   0]
print("Canal Blue:", cores['B'])   # [  0   0 255   0   0 128]
print("Canal Alpha:", cores['A'])  # [255 255 255   0 128 255]

# Filtrar cores opacas (alpha = 255)
opacas = cores[cores['A'] == 255]
print("\nCores opacas:", opacas)

# Modificar um campo
cores['A'][3] = 255  # Tornar a cor transparente em opaca
print("\nCor modificada:", cores[3])
````

````python
matriz_cores = np.zeros((5, 5), dtype=cor_rgba)

matriz_cores['R'] = np.random.randint(0, 256, (5, 5))
matriz_cores['G'] = np.random.randint(0, 256, (5, 5))
matriz_cores['B'] = np.random.randint(0, 256, (5, 5))
matriz_cores['A'] = 255  # Todas opacas

print("Matriz 5x5 de cores RGBA:")
print(matriz_cores)

print("\nPixel (0,0):", matriz_cores[0, 0])
print(f"R: {matriz_cores['R'][0,0]}")
print(f"G: {matriz_cores['G'][0,0]}")
print(f"B: {matriz_cores['B'][0,0]}")
print(f"A: {matriz_cores['A'][0,0]}")
````
